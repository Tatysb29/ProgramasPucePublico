### **Glosario de trminos de biologa computacional:**

**Algoritmo:** Es un procedimiento que se realiza paso a paso para conseguir un fin determinado.

**Fasta:** Es un formato simple que codifica un identificador, una descripcin de texto libre y la secuencia de cada protena. El formato es muy simple, utilizado por la mayora de los motores de bsqueda y herramientas de procesamiento posterior, y casi todos los proveedores de listas de secuencias de protenas lo exportan

**Motif (motivos):** Son regiones cortas y conservadas de una secuencia peptdica o de cido nucleico. En las protenas, los motivos se pueden delinear a nivel de la estructura primaria, secundaria y terciaria.

**PCC:** Es un coeficiente de correlacin lineal tambin conocido como PCC que es ampliamente usado para medir la asociacin entre pares de valores, es un coeficiente  muy usada para evaluar el desempeo de las matrices de peso.

***Matriz de puntuacin especfica de posicin (PSSM): ***Es una tabla de puntuacin especfica de la posicin derivada de una regin conservada de un alineamiento de secuencias y utilizada para caracterizar y clasificar secuencias segn su pertenencia a diferentes familias de genes o dominios. En contraste con los enfoques de diagnstico basados en motivos, que se enfocan en regiones locales de gran similitud, los perfiles generalmente abarcan familias o dominios de genes en toda su longitud, incluidas las regiones con gaps. El sistema de puntuacin define qu residuos se permiten en determinadas posiciones; qu posiciones se conservan y cules degeneran; y qu regiones pueden tolerar las inserciones. 

**Red neuronal artificial:** Es un modelo computacional evolucionado que presenta un conjunto de unidades, llamadas neuronas artificiales, las cuales estan conectadas entre s para transmitir seales. La informacin de entrada atraviesa la red neuronal (donde se somete a diversas operaciones) produciendo unos valores de salida. Cada neurona est conectada con otras a travs de unos enlaces, en estos enlaces el valor de salida de la neurona anterior es multiplicado por un valor de peso. Estos pesos en los enlaces pueden incrementar o inhibir el estado de activacin de las neuronas adyacentes. Del mismo modo, a la salida de la neurona, puede existir una funcin limitadora o umbral, que modifica el valor resultado o impone un lmite que no se debe sobrepasar antes de propagarse a otra neurona.
