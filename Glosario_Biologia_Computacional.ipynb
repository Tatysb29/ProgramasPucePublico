### **Glosario de términos de biología computacional:**

**Algoritmo:** Es un procedimiento que se realiza paso a paso para conseguir un fin determinado.

**Fasta:** Es un formato simple que codifica un identificador, una descripción de texto libre y la secuencia de cada proteína. El formato es muy simple, utilizado por la mayoría de los motores de búsqueda y herramientas de procesamiento posterior, y casi todos los proveedores de listas de secuencias de proteínas lo exportan

**Motif (motivos):** Son regiones cortas y conservadas de una secuencia peptídica o de ácido nucleico. En las proteínas, los motivos se pueden delinear a nivel de la estructura primaria, secundaria y terciaria.

**PCC:** Es un coeficiente de correlación lineal también conocido como PCC que es ampliamente usado para medir la asociación entre pares de valores, es un coeficiente  muy usada para evaluar el desempeño de las matrices de peso.

***Matriz de puntuación específica de posición (PSSM): ***Es una tabla de puntuación específica de la posición derivada de una región conservada de un alineamiento de secuencias y utilizada para caracterizar y clasificar secuencias según su pertenencia a diferentes familias de genes o dominios. En contraste con los enfoques de diagnóstico basados en motivos, que se enfocan en regiones locales de gran similitud, los perfiles generalmente abarcan familias o dominios de genes en toda su longitud, incluidas las regiones con gaps. El sistema de puntuación define qué residuos se permiten en determinadas posiciones; qué posiciones se conservan y cuáles degeneran; y qué regiones pueden tolerar las inserciones. 

**Red neuronal artificial:** Es un modelo computacional evolucionado que presenta un conjunto de unidades, llamadas neuronas artificiales, las cuales estan conectadas entre sí para transmitir señales. La información de entrada atraviesa la red neuronal (donde se somete a diversas operaciones) produciendo unos valores de salida. Cada neurona está conectada con otras a través de unos enlaces, en estos enlaces el valor de salida de la neurona anterior es multiplicado por un valor de peso. Estos pesos en los enlaces pueden incrementar o inhibir el estado de activación de las neuronas adyacentes. Del mismo modo, a la salida de la neurona, puede existir una función limitadora o umbral, que modifica el valor resultado o impone un límite que no se debe sobrepasar antes de propagarse a otra neurona.
